```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
data = pd.read_csv('/content/BuyerRatio.csv')
data
```

# Chi square test
Ho : no differenece in buyer ratios of male & female

H1 : significant difference in buyer ratios of male & female

```python
ratio = np.array([[50,142,131,70],[435,1523,1356,750]])
ratio
```

```python
chi2_contingency(ratio) # returns chi square stats, p, Dof, observed val
```

Since the (pvalue=0.66) > (alpha=0.5), therefore accept the null hypothesis.

***Inference : There is no difference in buyer ratios of male & female across the regions***
